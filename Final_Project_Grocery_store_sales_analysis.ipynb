**1.What is the percentage of female and male customers?**

```python
#count the number male and female customers
data = df.Gender.value_counts(ascending=True)
data
```

```python
# create a pie chart 
fig, ax = plt.subplots(figsize=(4, 4), subplot_kw=dict(aspect="equal"))
data.plot(kind='pie', startangle=90, autopct='%1.1f%%', colors=['#8d0047', '#00538d']);
ax.set_title('Customers by Genger')
ax.legend(title="Gender",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.show(); 
```

There are 17,1% Male and 82,9% Female customers 

**2.How many purchases were made by men and women for different product categories?**

```python
invoices_by_category_and_gender = df.pivot_table(index=['Gender'], columns=['Category'], values=['Receipt Number'], aggfunc='count')
display(invoices_by_category_and_gender)
```

**3.Identify the top 5 categories that are most purchased by customers based on gender**

```python
#Establish new user groups: Male/Female customers  
df_male = df[df['Gender'] == 'Male']
df_female = df[df['Gender'] == 'Female']
```

```python
male_category = df_male['Category'].value_counts()[:5]
male_category
```

```python
male_category_name = ['Bakery & Breakfast', 'Frozen', 'Beauty', 'Fruit & Vegetable', 'Pet Care'] 
```

```python
female_category = df_female['Category'].value_counts()[:5]
female_category
```

```python
female_category_name = ['Bakery & Breakfast', 'Frozen', 'Beauty', 'Fruit & Vegetable', 'Pet Care'] 
```

```python
plt.figure(figsize=(16,5))
base_color = sb.color_palette()[0]
ax1 = plt.subplot(1,2,1)
male_category.plot(kind='bar', color = base_color)
plt.xticks([0,1,2,3,4],male_category_name,rotation=15)
plt.title('Male customers by Category')
plt.xlabel('Top 5 Category')
plt.ylabel('Quantity')


ax2 = plt.subplot(1,2,2)
female_category.plot(kind='bar', color = base_color)
plt.xticks([0,1,2,3,4], female_category_name, rotation=15)
plt.title('Female customers by Category')
plt.xlabel('Top 5 category')
plt.ylabel('Quantity')
plt.show();
```

This analysis reveals that the top categories for both genders remain the same - Bakery & Breakfast', 'Frozen', 'Beauty', 'Fruit & Vegetable', 'Pet Care'. The only differences are in quantity.

**6.Find out the Gross Bill Amount by year for each country. I'm wondering if the Gross Bill has increased**

```python
# check what and how many years do we have in columns
years = df['Year'].unique()
print(years)
```

```python
gross_country_pivot = df.pivot_table(index = 'Year', columns = 'Country', values = 'Gross Bill Amount', aggfunc= 'sum')
gross_country_pivot.plot()
plt.title('Dynamic chart showing Gross Bill Amount from 2016 to 2019')
plt.xlabel('Year')
plt.ylabel('Gross Bill Amount')
plt.show();
```
